```python
# Import all libraries
import pandas as pd
import sklearn.datasets as datasets
```

```python
# Load DataSet
iris=datasets.load_iris()

# Forming the iris dataframe
df=pd.DataFrame(iris.data, columns=iris.feature_names)
print(df.head(5))
```

```python
# Print target
y=iris.target
print(y)
```

**DEFINE THE DECISION TREE ALGO**

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
clf.fit(df,y)

print('Decision Tree Classifer Created')
```

**VISUALIZE THE DECISION TREE**

```python
# Install required libraries
!pip install pydotplus
!apt-get install graphviz -y
```

```python
# Import necessary libraries for graphviz
from six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

# Visualize the graph
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data, feature_names=iris.feature_names, filled=True, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

**Now we can feed any new/test data to this classifer and it would be able to predict the right class accordingly.**
